```python
from gtts import gTTS
import os

# A korábban definiált magyar szöveg
# hungarian_text = "..."

# Létrehozzuk a gTTS objektumot
tts = gTTS(text=hungarian_text, lang='hu', slow=False)

# Elmentjük az audio fájlt
audio_file_path = "magyar_szoveg.mp3"
tts.save(audio_file_path)

print(f"A szöveg sikeresen hangfájlként elmentve ide: {audio_file_path}")
```
